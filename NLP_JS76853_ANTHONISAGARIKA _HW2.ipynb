```python
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
def filter (text):
    stop_words =set(stopwords.words('english'))
    words = [word.lower() for word in text if word.isalpha() and word.lower() not in stop_words]
    fd = nltk.FreqDist(words)
    return fd.most_common(50)

reuters_word = nltk.corpus.reuters.words()
result = filter(reuters_word)
print(result)
```

#Confirming Zip's law

Since Zip's law is almost related to logarithmic functions I would say because of the following reasons

```python
import pylab
import numpy as np
```

```python
def plot_word_freq():
    text = brown.words(categories = 'science_fiction')
    words = [word.lower() for word in text if word.isalpha()]
    unique_words, counts = np.unique(words,return_counts=True)
    sorted_counts = np.sort(counts)[::-1]
    ranks = np.arange(1,len(sorted_counts)+1)
    pylab.figure(figsize=(10,6))
    pylab.xlabel('Rank')
    pylab.ylabel('Frequency')
    pylab.plot(ranks,np.log(sorted_counts),marker = 'o')
    pylab.show()
   

plot_word_freq()
```

As zip law states there is extreme right end and there are apparent steps inbetween that means there are often occurances which are low and at the right end it is flat which mean the frequencies are almost similar and they donot drop quite often which states Zip's law

```python
import random
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
test = [random.choice("abcdefghijklmnopqrstuvwxyz ") for _ in range(80000)]

# Count word frequencies
word_freq = {}
for word in test:
    word_freq[word] = word_freq.get(word, 0) + 1

# Sort word frequencies
sorted_word_freq = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)

# Extract frequencies and ranks
frequencies = [item[1] for item in sorted_word_freq]
ranks = np.arange(1, len(frequencies) + 1)

# Plot Zipf's law
plt.plot(np.log(ranks), np.log(frequencies))
plt.xlabel('log(Rank)')
plt.ylabel('log(Frequency)')
plt.title("Zipf's Law")
plt.show()
```

```python
#Using udhr.corpus for to identify the given where it appears in different languages

#searching for latin files


def find_language(string):
    #using two lists for the given list of languages and the resulted ones
    languages = []
    required_languages = []
    #creating a for loop for reading the names in the udhr
    for name in nltk.corpus.udhr.fileids():
        if 'Latin1' in name:
            #adding the name into list
            languages.append(name)
    for lang in languages:
        if string in nltk.corpus.udhr.words(lang):
            #adding the output into the new list
            required_languages.append(lang)
    return required_languages

#Example :

languages = find_language("rights")
print(languages)

```
