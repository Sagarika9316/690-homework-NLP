First we are going to use affixtagger to the brown corpus later we are going to tokenize the new sentence and  train the sentence


```python
from nltk.corpus import brown
brown_tagged_sents = brown.tagged_sents(categories = 'news')
brown_sents = brown.sents(categories = 'news')
# using an affix tagger to train the brown.corpus
affix_tagger = nltk.AffixTagger(brown_tagged_sents)
#Training the new sentence
tagger_sentence = "This a test sentence and I know this sentence can help me getting my expected test and train case scenarios"
tokenization = nltk.word_tokenize(tagger_sentence)
new_tagged_sentence = affix_tagger.tag(tokenization)
#Tagging the new sentence if it is present in brown.corpus()
print(new_tagged_sentence)
```

The above sentence returns None in some cases because this word hasnt been affix tagged in the brown corpus

```python
#Experimenting with different affix_lengths
affix_length = [2,3,4]
minimum_lengths = [3,4,5]
for affix_len in affix_length:
    for min_word_length in minimum_lengths:
        affix_tagger = nltk.AffixTagger(brown_tagged_sents,affix_length=affix_len,min_stem_length = min_word_length)
        accuracy = affix_tagger.evaluate([new_tagged_sentence])
        print(f"Accuracy with {affix_len} and {min_word_length} is", accuracy)
```

```python
from nltk.corpus import brown
def find_tags(tagged_text):
    #using nltk freq dist to find the word is noun and nouns ending with 's'
    singular_nouns = nltk.FreqDist(word.lower() for (word,tag)in tagged_text if tag == 'NN' and not any(char.isdigit() for char in word))
    plural_nouns = nltk.FreqDist(word.lower() for (word,tag) in tagged_text if tag == 'NNS' and not any(char.isdigit() for char in word))
    #checking if the plurals are repeated more than once
    common_plural_nouns = {word : freq for word,freq in plural_nouns.items() if freq > 1   }
    return common_plural_nouns

```

```python
#Counting with example
```

```python
tagged_text = brown.tagged_words()
plural_nouns = find_tags(tagged_text)

for noun,freq in plural_nouns.items():
    print(noun,freq)
```

Which words have greatest number of distinct tags?What are they? and what do they represent?

```python
brown_news_tagged = brown.tagged_words()
data = nltk.ConditionalFreqDist((word.lower(),tag)
                                for(word,tag) in brown_news_tagged)
for word in sorted(data.conditions()):
  if len(data[word])> 3:
      tags = [tag for (tag,_) in data[word].most_common()]
      #using set to find distinct tags
      distinct_tags = set(tags)
      print(word,' '.join(tags),len(distinct_tags))
      
```

```python
#Which tags are nouns most commonly found after ? what do these tags represent?
```

```python
from collections import defaultdict


pos = defaultdict(lambda : defaultdict(int))
brown_news_tagged = brown.tagged_words()
for((w1,t1),(w2,t2)) in nltk.bigrams(brown_news_tagged):
    if t1.startswith('N'):
        pos[t2][t1] +=1


for tag,next_tags in pos.items():
    most_tag = max(next_tags,key = next_tags.get)
    print(f"Tag : {tag},Most common nountag: {most_tag},count :{next_tags[most_tag]}")
```

From the above code we can see that the Tag:IN has more counts followed by  a Comma , tag

Most commonly found tags before noun tags are IN "preposition" so we can say after prepositions the nouns usually occur
